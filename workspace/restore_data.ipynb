```python
"""Utilities for restoring deleted BigQuery datasets and tables."""
from google.cloud import bigquery

client = bigquery.Client()
```

```python
query = """
## [Restore deleted dataset](https://docs.cloud.google.com/bigquery/docs/restore-deleted-datasets#restore_a_dataset)
UNDROP SCHEMA httparchive.crawl;
"""
client.query(query).result()
```

```python
# [Restore deleted table]
# https://docs.cloud.google.com/bigquery/docs/restore-deleted-tables#restore_a_table
!date -d '2025-08-04 16:00:00.000000Z' +%s000
!bq cp httparchive.crawl.pages@1759670400000 httparchive.crawl_staging.pages_restored_20250804
```

```python
query = """
## [Restore a table to a specific point in time](https://cloud.google.com/bigquery/docs/restore-tables#restoring_a_table_to_a_specific_point_in_time)
CREATE TABLE httparchive.crawl_staging.pages_restored_20250804 AS
SELECT *
FROM httparchive.crawl.pages
  FOR SYSTEM_TIME AS OF TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL 1 HOUR);
"""
client.query(query).result()
```
